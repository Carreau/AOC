```python
data = """............
........0...
.....0......
.......0....
....0.......
......A.....
............
............
........A...
.........A..
............
............"""

data = Path('2024-8.txt').read_text()

grid = [[s for s in line ] for line in data.splitlines()]
for l in grid :
    print(''.join(l))
```

```python
S = len(grid)
```

```python
towers = defaultdict(list)
for i,line in enumerate(grid):
    for j,s in enumerate(line):
        if s != '.':
            towers[s].append((i,j))
towers
```

```python
nodes = defaultdict(list)
for k,tows in list(towers.items()):
    for (tay,tax),(tby, tbx) in list(product(tows,tows)):
        if tax == tbx and tay == tby:
            continue
        kx,ky = (2*tax-tbx, 2*tay-tby)
        
        if 0<=kx<S and 0<=ky<S:
            nodes[(ky,kx)].append(k)
        else:
            print(ky,kx)
    
```

```python
from copy import deepcopy
gx = deepcopy(grid)
c = 0
for y,x in nodes.keys():
    assert gx[y][x] != '#'
    gx[y][x] = '#'
    c+=1

for l in gx:
    print(''.join(l))
```

```python
nodes = defaultdict(list)
for k,tows in list(towers.items()):
    for (tay,tax),(tby, tbx) in list(product(tows,tows)):
        if tax == tbx and tay == tby:
            #print('same tower')
            continue
        dx,dy = tbx-tax, tby-tay
        #print(dx,dy)
        g = gcd(dx,dy)
        dx,dy == dx /g, dy/g
        if g != 1:
            print(dx,dy ,'by', g)

        
        for u in range(-S,S):
            kx,ky = (tax-u*dx, tay-u*dy)
            print(kx,ky)
            if 0<=kx<S and 0<=ky<S:
                nodes[(ky,kx)].append(k)
```
