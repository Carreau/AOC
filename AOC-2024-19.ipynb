```python
from pathlib import Path
from functools import lru_cache, cache

```

```python
data = """r, wr, b, g, bwu, rb, gb, br

brwrr
bggr
gbbr
rrbgbr
ubwu
bwurrg
brgr
bbrgwb"""
data = Path('19').read_text()

towels, patterns = data.split('\n\n')
towels = tuple(towels.split(', '))
patterns = patterns.splitlines()
towels, patterns
```

```python
@cache
def can_make(seq, towels):
    if not seq:
        return True
    for t in towels:
        if seq == t:
            return True
        elif seq.startswith(t):
            res = can_make(seq[len(t):], towels)
            if res:
                return True

    return False
    
```

```python
sum([can_make(s, towels) for s in patterns])
```

```python
@cache
def arrangements(seq, towels):
    acc = 0
    for t in towels:
        if seq == t:
            acc += 1
        elif seq.startswith(t):
            acc += arrangements(seq[len(t):], towels)
    return acc
```

```python
sum([arrangements(s, towels) for s in patterns]) # 809 too low
```
