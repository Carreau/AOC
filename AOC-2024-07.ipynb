```python
def _all_pos(acc, numbers):
    if len(numbers) == 1:
        n0 = numbers[0]
        yield acc*n0
        yield acc+n0
    else:
        if len(numbers) == 2:
            n0,ns = numbers
            ns = [ns]
        else:
            assert len(numbers) > 2, numbers 
            n0,*ns = numbers
        yield from _all_pos(acc*n0, ns)
        yield from _all_pos(acc+n0, ns)

def all_pos(numbers):
    n0,*ns = numbers
    return list(_all_pos(n0, ns))
        
```

```python
def _all_pos2(acc, numbers):
    if len(numbers) == 1:
        n0 = numbers[0]
        yield acc*n0
        yield acc+n0
        yield int(str(acc)+str(n0))
    else:
        if len(numbers) == 2:
            n0,ns = numbers
            ns = [ns]
        else:
            assert len(numbers) > 2, numbers 
            n0,*ns = numbers
        yield from _all_pos2(acc*n0, ns)
        yield from _all_pos2(acc+n0, ns)
        yield from _all_pos2(int(str(acc)+str(n0)), ns)

def all_pos2(numbers):
    n0,*ns = numbers
    return list(_all_pos2(n0, ns))
```
