```python
from operator import and_, or_, xor
```

```python
gates = {}
zs = []
outputwires = []
for l in _gates.splitlines():
    rule, out = l.split(' -> ')
    if out.startswith('z'):
        zs.append(out)
    outputwires.append(out)
    gates[out] = rule.split()



def get(name):
    map = {
        'AND': and_,
        'OR': or_,
        'XOR': xor
        
    }
    if name in data:
        return data[name]
    else:
        a,op,b = gates[name]
        return map[op](get(a),get(b))
    assert False
```

```python

def getb(name, x, y,gates):
    map = {
        'AND': and_,
        'OR': or_,
        'XOR': xor
    }
    if name.startswith('x'):
        return x[int(name[1:])]
    if name.startswith('y'):
        return y[int(name[1:])]
    else:
        a,op,b = gates[name]
        return map[op](getb(a, x, y, gates),getb(b, x,y, gates))
    assert False
```

```python
def swapgate(gates, swaps):
    g2 = {x:y for x,y in gates.items()}
    for sa,sb in swaps:
        if sa is None:
            continue
        g2[sa] = gates[sb]
        g2[sb] = gates[sa]
    return g2
        
        
    
```
