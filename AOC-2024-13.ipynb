```python
data = """Button A: X+94, Y+34
Button B: X+22, Y+67
Prize: X=8400, Y=5400

Button A: X+26, Y+66
Button B: X+67, Y+21
Prize: X=12748, Y=12176

Button A: X+17, Y+86
Button B: X+84, Y+37
Prize: X=7870, Y=6450

Button A: X+69, Y+23
Button B: X+27, Y+71
Prize: X=18641, Y=10279
""".split('\n\n')
from pathlib import Path

data = Path('2024-13.txt').read_text().split('\n\n')

machines = []

for d in data:
    a,b,p = d.splitlines()
    a = a.split(':')[1]
    b = b.split(':')[1]
    p = p.split(':')[1]
    ax,ay = [int(_.split('+')[1]) for _ in a.split(',')]
    bx,by = [int(_.split('+')[1]) for _ in b.split(',')]
    px,py = [int(_.split('=')[1]) for _ in p.split(',')]
    #print(ax*by - bx*ay)
    machines.append((ax,ay, bx,by, px,py))


#      ax bx| u     = px
#      ay by| v     = py
```

```python
import numpy as np
from numpy.linalg import inv
```

```python
count = 0
for ax,ay,bx,by,px,py in machines:
    A = np.array([[ax,bx],[ay,by]])
    P = np.array([px,py])
    Am = inv(A)
    _ma,_mb = Am@P
    ma, mb = int(np.round(_ma)), int(np.round(_mb))    
    cand = np.array([int(ma), int(mb)])
    #print(ma,mb, cand)
    if (A@cand == P).all():
        cost = 3*ma+mb
        #print(cand, cost)
        count += cost
    else:
        print(_ma,_mb, ma,mb)
count
```
