```python
import numpy as np

data = """#####
.####
.####
.####
.#.#.
.#...
.....

#####
##.##
.#.##
...##
...#.
...#.
.....

.....
#....
#....
#...#
#.#.#
#.###
#####

.....
.....
#.#..
###..
###.#
###.#
#####

.....
.....
.....
#....
#.#..
#.#.#
#####"""

from pathlib import Path
data = Path('25.txt').read_text()

_schmematics = data.split('\n\n')
len(_schmematics)
```

```python
_keys = []
_locks = []
for s in _schmematics:
    korl = (np.array([list(l) for l in s.splitlines()]) == '#')
    if korl[0,0]:
        _locks.append(korl)
    else:
        _keys.append(korl)
```

```python
locks = [l.sum(axis=0)-1 for l in _locks]
keys = [k.sum(axis=0)-1 for k in _keys]
```

```python
len(locks), len(keys)
```

```python
def nfit(keys, locks):
    n =0
    for k in keys:
        for l in locks:
            if (k+l).max() <=5:
                n+=1
    return n

nfit(keys, locks)
```
