```python
def base_safe(levels):
    arr = np.array(levels)
    
    diff = arr[1:]-arr[:-1]
    return ((diff > (-4)) & (diff < 0)).all() | ((diff < (4)) & (diff  > 0)).all()
```

```python
n = 0
for levels in acc:
    if base_safe(levels):
      n+=1
n
```

```python
from copy import copy
```

```python
n = 0
m = 0 
for levels in acc:
    if base_safe(levels):
        n+=1
        continue
    else:
        for i in range(len(levels)):
            lc = copy(levels)
            lc.pop(i)
            if base_safe(lc):
                m+=1
                break
        
    
n,m , n+m
```
