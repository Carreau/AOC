```python
def ann(ax, data=data):
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            text = ax.text(j, i, data[i, j],
                       ha="center", va="center", color="w")
```

```python
from itertools import product
scenic = np.zeros_like(data)
WD = data.shape[0]
for j,i in product(range(WD), range(WD)):
    
    treeh = data[j,i]
    print('== ',j,i, '->', treeh, '==')
    up, left, right, down = 1,1,1,1
    if i in (0,WD-1) or j in (0,WD-1):
        scenic[j,i] = 0
        continue
    for y in range(j-1,-1,-1):
        if data[y,i] >= treeh or y == 0:
            print('up on', y,i, data[y,i])
            up = j-y
            break
    

    for y in range(j+1,WD):
        if data[y,i] >= treeh or y == WD-1:
            print('down on', y,i, data[y,i])
            down = y-j
            break

    
    for x in range(i-1,-1,-1):
        if data[j,x] >= treeh or x == 0:
            print('left on', j,x, data[j,x])
            left = i-x
            break

    for x in range(i+1,WD):
        if data[j,x] >= treeh or x == WD-1:
            print('right on', j,x, data[j,x])
            right = x-i
            break

    scenic[j,i] = up*left*right*down
    print(treeh, scenic[j,i],'=', up, left, right, down)
        
```
