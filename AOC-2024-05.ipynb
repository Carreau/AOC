```python
from pathlib import Path
from collections import defaultdict

data = Path('2024-05.txt').read_text()


# data = """47|53
# 97|13
# 97|61
# 97|47
# 75|29
# 61|13
# 75|53
# 29|13
# 97|29
# 53|29
# 61|53
# 97|53
# 61|29
# 47|13
# 75|47
# 97|75
# 47|61
# 75|61
# 47|29
# 75|13
# 53|13

# 75,47,61,53,29
# 97,61,53,29,13
# 75,29,13
# 75,97,47,61,53
# 61,13,29
# 97,13,75,29,47
# """
order, sequences = data.split('\n\n')
sequences = [s.split(',') for s in sequences.splitlines()]
```

```python
pairs = [p.split('|') for p in order.splitlines()]
dorder = defaultdict(list)
for a,b in pairs:
    dorder[a].append(b)
dorder
```

```python
def test_seq(seq):
    for i,n in enumerate(seq):
        for m in seq[i:]:
            if n in dorder.get(m,[]):
    #            print('bad sequence')
                return False
    #print('Good')
    return True
            
    
sn = [seq for seq in sequences if test_seq(seq)]
sum([int(seq[len(seq)//2]) for seq in sn])
    
    
            
```
