```python
def move_block_id(fs,id):
    indexes = [b.id for b in fs]
    ix = indexes.index(id)
    tomove = fs[ix]
    free = [i for i,b in enumerate(fs) if b.empty >= tomove.occupied and i < ix ]
    if not free:
        return fs
    freex = free[0]
    fs[ix] = Block(0, tomove.occupied+tomove.empty, None)
    a,b,c = fs[:freex],fs[freex],fs[freex+1:]

    fs = a + [Block(b.occupied, 0, b.id), Block(tomove.occupied, b.empty-tomove.occupied, tomove.id  )] + c
    return fs
    
nfs = copy(fs)

for i in range(len(fs)-1, 0, -1):
    #print(''.join(str(s) for s in nfs))
    nfs = move_block_id(nfs, i)
    
```
