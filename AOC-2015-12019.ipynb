```python
def mpart(mol):
    assert isinstance(mol,str)
    lmol = split(mol)
    
    # 4 parts,
    acc = []
    cacc = ''
    count = 0
    for atom in lmol:
        if atom == 'Rn':
            count +=1
            cacc += atom
        elif atom == 'Ar':
            count-=1
            cacc += atom
            if count == 0:
                acc.append(cacc)
                cacc = ''
        else:
            cacc += atom
    if cacc:
        acc.append(cacc)
    return acc
            
def iter_all(mol:str):
    yield mol
    for k,v in backward.items():
        if k in mol:
            nm =  mol.replace(k, v, 1)
            yield nm
            yield from set(iter_all(nm))
            
            
def final(mol):
    print('FInal', mol)
    assert isinstance(mol, str)
    lmol = split(mol)
    assert lmol.count('Ar') == 1
    assert lmol.count('Rn') == 1
    a,b = mol.split('Rn')
    _as = set(iter_all(a))
    _bs = set(iter_all(b))
    return [''.join(x) for x in product(_as, ['Rn'], _bs)]
    
def rec(mol):
    print('Rec', mol)
    lmol = split(mol)
    if lmol.count('Ar') == 1:
        yield from final(mol)
    else:
        assert 'Rn' in mol, mol
        a,_b = mol.split('Rn', maxsplit=1)
        b = _b[:-2]
        parts = mpart(b)
        _as = iter_all(a)
        _bs = rec(b)
        yield from [''.join(x) for x in product(_as,['Rn'],_bs,['Ar'])]


a,b,c,d,e,f = mpart(mol)
b,set(rec(a))
```
