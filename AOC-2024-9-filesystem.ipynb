```python
data = '2333133121414131402'
from pathlib import Path
data = Path('2015-11.txt').read_text()[:-1]
```

```python
from itertools import batched, cycle
```

```python
i = 0 
acc = []
for run, block in zip(data, cycle([True,False])):
    if block:
        acc.extend([i]*int(run))
        i+=1
    else:
        acc.extend(['.']*int(run))
''.join([str(i) for i in acc])
```

```python
i= 0 
while i < len(acc):
    if acc[i] == '.':
        last = acc.pop()
        while last == '.':
            last = acc.pop()
        acc[i] = last
    i+=1
    #print(''.join([str(s) for s in acc]))
```
