```python
def search(start):
    closed = set()
    open = {start}
    while open:
        (cy,cx), cheight = open.pop()
        closed.add(((cy,cx), cheight))
        neighboor = [((cy+i,cx+j),tmap.get((cy+i,cx+j), 200)) for i,j in [(-1,0),(1,0), (0,1), (0,-1)]]
        neighboor = {((y,x),height) for ((y,x),height) in neighboor if height == cheight+1}
        open.update(neighboor)
        open.difference_update(closed)
    return closed

len([s for s in search(start) if s[1] == 9])
```
