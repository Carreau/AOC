```python

smallgals  = list(zip(*np.where(sky_c)))

empty_rows = np.where(~rows)
empty_cols = np.where(~cols)
print(empty_rows)
print(empty_cols)
N = 1_000_000
from itertools import permutations
tot = 0
plt.matshow(sky_c)   
for i in range(len(smallgals)):
    for j in range(i+1,len(smallgals)):
        g1 = smallgals[i]
        g2 = smallgals[j]
        y,Y = sorted((g1[0], g2[0]))
        x,X = sorted((g1[1], g2[1]))
        xspaces = ((x < empty_cols) & (empty_cols < X)).sum()
        yspaces = ((y < empty_rows) & (empty_rows < Y)).sum()
        distancey = (Y-y)-yspaces + yspaces*N 
        distancex = (X-x)-xspaces + xspaces*N
        tot +=(distancey+distancex)
     
tot # 458191688761
```
