```python

def explore(current, dir):
    visited = [current]
    for i in range(20164):
        new_pos = current + dir
        visited.append(new_pos)
        possible_new_dir = [x for x in dirs[maze[*new_pos]] if (x != -dir).any()]
        if not possible_new_dir:
            return None, None
        if len(possible_new_dir) == 1:
            #print('going in only possible direction', possible_new_dir[0])
            next_dir = possible_new_dir[0]
        else:
            if maze[*new_pos] == 'S':
                return visited,i
            print('we encountered', len(possible_new_dir), 'directions...', maze[*new_pos])
            
            return None, None
        current = new_pos
        dir = next_dir
    return [coord, dir, i]



for d in dirs[maze[*start_coord]]:
    vs, i = explore(start_coord,  d)
    if vs:
        print((i+1) // 2)
    if vs is not None:
        visited = vs
```
