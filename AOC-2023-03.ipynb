```python
import re
import string
partss = set(''.join(data)) - set(string.digits) - set('.')
partss

```

```python
mdata = ['.'*len(data[0])]+['.'+d+'.' for d in data]+['.'*len(data[0])]
```

```python
spans = []
for i,line in enumerate(mdata):
    for m in re.finditer('\d+', line):
        
        spans.append([i, m.span()])
```

```python
tot = 0
ptot = []
c = 1
for i,(x,y) in spans:
    if i != c:
        print(f'line {i-2}: {ptot}')
        c = i
        ptot = []
    number = mdata[i][x:y]
    
    ispart = bool(partss.intersection(set(mdata[i-1][x-1:y+1]+mdata[i][x-1:y+1]+mdata[i+1][x-1:y+1])))
    if ispart:
        tot += int(number)
        ptot.append(int(number))
    if not ispart:
        print(number, 'not accounted')

    #print(number, ispart)
    
    
tot    # 527369
```
